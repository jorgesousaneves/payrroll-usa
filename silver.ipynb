```python
from pyspark.sql.functions import year, to_date, col

df_bronze = spark.table("payroll.bronze.eua_payroll_raw")

df_silver = df_bronze.select(
    to_date(col("data_referencia_str"), "yyyy-MM-dd").alias("data_referencia"),
    col("variacao_liquida_payroll_milhares"),
    year(to_date(col("data_referencia_str"), "yyyy-MM-dd")).alias("ano") 
)

df_silver = df_silver.filter(col("data_referencia").isNotNull())

df_silver.write.format("delta") \
 .mode("overwrite") \
 .option("mergeSchema", "true") \
 .partitionBy("ano") \
 .saveAsTable("payroll.silver.eua_payroll_tratado")
```
