```python
%sql
USE CATALOG payroll;
USE SCHEMA gold;

CREATE OR REPLACE TABLE ft_payroll_evolucao_eua
USING DELTA
AS
SELECT
    data_referencia,
    DATE_FORMAT(data_referencia, 'yyyy-MM') AS ano_mes_chave,
    variacao_liquida_payroll_milhares AS criacao_liquida_empregos_eua_milhares
FROM payroll.silver.eua_payroll_tratado

ORDER BY data_referencia ASC; 
```
